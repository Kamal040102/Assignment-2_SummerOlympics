```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv("Summer.csv")
```

# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
data = []

for city in df['City'].unique():
    data.append(city)
    
len(data)
```

From the above o/p we can say that Summer Olympics has been help in 22 cities so far.

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
data = []

for Sport in df['Sport'].unique():
    data.append([Sport, len(df.loc[(df['Medal']=='Gold') & (df['Sport']==Sport)])])
    
pd.DataFrame(data,columns=['Sport','numberOfGoldMedal']).sort_values(by='numberOfGoldMedal', ascending=False).head().plot.bar(x='Sport',y='numberOfGoldMedal',figsize=(7,7))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data=[]

for Sport in df['Sport'].unique():
    data.append([Sport,len(df[df['Sport']==Sport])])
    
pd.DataFrame(data,columns=['Sport','Medals']).sort_values(by='Medals',ascending=False).head().plot.bar(x='Sport',y='Medals',figsize=(7,7))
```

### 4. Which player has won most number of medals? (Top 5)

```python
data = []

for Athlete in df['Athlete'].unique():
    data.append([Athlete,len(df[df['Athlete']==Athlete])])
    
pd.DataFrame(data,columns=['Athlete','numberOfMedal']).sort_values(by='numberOfMedal',ascending=False).head().plot.bar(x='Athlete',y='numberOfMedal',figsize=(7,7))
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
data = []

for Athlete in df['Athlete'].unique():
    data.append([Athlete,len(df.loc[(df['Athlete']==Athlete) & (df['Medal']=='Gold')])])
    
pd.DataFrame(data,columns=['Athlete','numberOfMedal']).sort_values(by='numberOfMedal',ascending=False).head().plot.bar(x='Athlete',y='numberOfMedal',figsize=(7,7))
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
dict = []
data = []

for Year in df['Year'].unique():
    dict.append([Year,len(df.loc[(df['Year']==Year) & (df['Country']=='IND') & (df['Medal']=='Gold')])])
    
for i in range (len(dict)):
    for j in range(1,2,1):
        if dict[i][j] >= 1:
            data.append([dict[i][0]])
            
result = str(data[0])
result = result.replace('[','')
result = result.replace(']','')

print(f"'IND' for the first time won gold medal in the year {result}.")
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data = []

for Event in df['Event'].unique():
    data.append([Event, len(df[df['Event']==Event])])
    
pd.DataFrame(data,columns=['Event','numberOfPlayer']).sort_values(by='numberOfPlayer',ascending=False).head().plot.bar(x='Event',y='numberOfPlayer',figsize=(7,7))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
data = []

for Sport in df['Sport'].unique():
    data.append([Sport, len(df.loc[(df['Sport']==Sport) & (df['Gender']=='Women') & (df['Medal']=='Gold')])])
    
pd.DataFrame(data,columns=['Sport','numberFemaleAthlete']).sort_values(by='numberFemaleAthlete', ascending=False).head().plot.bar(x='Sport',y='numberFemaleAthlete',figsize=(7,7))
```
